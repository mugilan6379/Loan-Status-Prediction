```python
import numpy as np 
import altair as alt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn import svm
```

```python
loan_dataset = pd.read_csv('loanPrediction.csv')
loan_dataset.head(5)
```

```python
loan_dataset.isnull().sum()
```

```python
loan_dataset.groupby(['Loan_Status','Married']).mean()
```

```python
loan_dataset=loan_dataset.dropna()
```

```python
loan_dataset.replace({'Loan_Status':{'N':0,'Y':1}},inplace=True)
```

```python
loan_dataset.head()
```

```python
loan_dataset.replace({'Dependents':{'3+':4}},inplace=True)
loan_dataset['Dependents'].value_counts()
```

```python
chart= alt.Chart(loan_dataset).mark_bar().encode(
    x='Education',
    y='count()',
    color='Education',
    column='Loan_Status'
).properties(title='Education vs Loan Status')
chart.show()
```

```python
chart= alt.Chart(loan_dataset).mark_bar().encode(
    x='Dependents',
    y='count()',
    color='Dependents',
    column='Loan_Status'
).properties(title='Education vs Loan Status',width=100)
chart.show()
```

```python
chart= alt.Chart(loan_dataset).mark_bar().encode(
    x='Married',
    y='count()',
    color='Married',
    column='Loan_Status'
).properties(title='Education vs Loan Status')
chart.show()
```

```python
loan_dataset.replace({'Married':{'Yes':1,'No':0},
                     'Education':{'Graduate':1,'Not Graduate':0},
                     'Gender':{'Male':1,'Female':0},
                     'Self_Employed':{'Yes':1,'No':0},
                     'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2}}
                     ,inplace=True)

loan_dataset['Loan_ID']=loan_dataset['Loan_ID'].str.replace('LP','',regex=True)
loan_dataset.head()
```

```python
X=loan_dataset.drop(columns=['Loan_Status','Loan_ID'],axis=1)
Y=loan_dataset['Loan_Status']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train,Y_train)
```

```python
X_train_pred = classifier.predict(X_train)
X_test_pred = classifier.predict(X_test)
train_acc = accuracy_score(X_train_pred,Y_train)
test_acc = accuracy_score(X_test_pred,Y_test)

print('Training accuracy: ',train_acc)
print('Testing accuracy: ',test_acc)
```
